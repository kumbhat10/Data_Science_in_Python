# **matplotlib**

## **plt.plot**

```
# for simple bar charts, line charts, and scatterplots, it works pretty well.

from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis
plt.plot(years, gdp, color='green', marker ='o',linestyle = 'solid')
plt.title("Nominal GDP")   # add a title
plt.ylabel("Billions of $") # add y label
plt.xlabel("Year")
plt.show()
```

# **plt.bar**   Bar charts


```
from matplotlib import pyplot as plt
movies = ["3 Idiots" ,"PK","Murder","Gravity"]
num_oscars = [5,11,3,8]
 # bars are by default 0.8 width so we will add 0.1  to the left cordinates so that each bar is centered
xs = [x+1  for x,_ in enumerate(movies)]  # xs is cordinate of center of the bar width 
# print(xs)  # [1, 2, 3, 4]

plt.bar(xs,num_oscars)   # plot bar with xs as centre of bar on x-axis and num_oscars as bar height
plt.ylabel("No of oscars Nominee")
plt.title("Summary of oscar nomination")
plt.xlabel("Name of the movie")

plt.xticks([i+1 for i,_ in enumerate(movies)],movies) # location of x-cordinate to inset xlabel 
plt.show()
```

# **Histogram**

```
# A bar chart can also be a good choice for plotting histograms of bucketed numeric values, in order to visually explore how the values are distributed.

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
decile = lambda grade: grade // 10 * 10   ## round to nearest tens

from collections import Counter
histogram = Counter(decile(x) for x in grades)

# print(histogram.most_common())  # most common values [(80, 4), (70, 3), (90, 2), (0, 2), (100, 1), (60, 1)]
# print(histogram.keys())   # dict_keys([80, 90, 70, 0, 100, 60])
# print(histogram.values()) # dict_values([4, 2, 3, 2, 1, 1])

xs = [x for x in histogram.keys()]  # xs is the center of each bar

plt.bar(xs,histogram.values(), 8)  # 8 is the width of each bar
plt.axis([-5,105,0,5])             # x- axis range from -5 to 105 and y axis range from 0 to 5
plt.xticks([x*10 for x in range(11)])
plt.xlabel("Grades - decile")
plt.ylabel("No of Students")
plt.title("Distribution of grades of mid term")
plt.show()
```

# **plt.plot** Line Charts

```
from matplotlib import pyplot as plt
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error  = [x+y for x,y in zip(variance, bias_squared)]
# print(total_error)  # [257, 130, 68, 40, 32, 40, 68, 130, 257]
xs = [x for x,_ in enumerate(variance)]
# print(xs) # [0, 1, 2, 3, 4, 5, 6, 7, 8]

plt.plot(xs,variance, 'g-',label='Variance')
plt.plot(xs,bias_squared,'r-.',label='Bias squared')
plt.plot(xs, total_error,'b--',label='Total error')
plt.legend(loc=9)  # because we have assigned labels to each plot, we can get legend - loc = 9 means "top center"
plt.title("The bias Variance Trade off")
plt.xlabel("Model complexity")
plt.show()
```

# **plt.scatter** - Scatter Plots

```
from matplotlib import pyplot as plt
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)
# label each point 
for label, friend_count, minutes_count in zip(labels,friends,minutes):
  plt.annotate(label, xy=(friend_count, minutes_count), xytext = (5,-5), textcoords='offset points')

plt.title("Daily  Minutes vs Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("Daily minutes spent on site")
# plt.axis("equal")   # make the range on both the axis same - but different min and max values - use it to avoid misleading plots with misleading axis range
plt.show()
```
